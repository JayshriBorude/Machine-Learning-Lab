```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("C:/Users/userp/Downloads/used_cars_data.csv")

# Clean 'Mileage' if it's in string format
data['Mileage'] = data['Mileage'].astype(str).str.extract(r'(\d+\.?\d*)').astype(float)

# Select top 5 cars by 'Kilometers_Driven'
top_cars = data[['Mileage', 'Kilometers_Driven']].sort_values(by='Kilometers_Driven', ascending=False).head(5)

# Plot
plt.scatter(top_cars['Mileage'], top_cars['Kilometers_Driven'], color='pink')
plt.xlabel('Mileage')
plt.ylabel('Kilometers Driven')
plt.title('Mileage of Cars by Kilometers Driven')

# Save the figure
plt.savefig(r"C:\Users\userp\Desktop\MLLabs.jpg")
plt.show()
```
